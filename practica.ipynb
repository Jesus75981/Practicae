Ejercicio 1

```python
import cv2
import matplotlib.pyplot as plt

img_original = cv2.imread('imagen1.jpg')

x1 = 650
y1 = 600
x2 = 850
y2 = 800

img_recortada = img_original[y1:y2, x1:x2]

cv2.imwrite('Imagenrecortada.jpg', img_recortada)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('on')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img_recortada, cv2.COLOR_BGR2RGB))
plt.title('Imagen Recortada')
plt.axis('on')

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

def generar_degradado_lineal(altura, ancho):
  
    degradado = np.zeros((altura, ancho, 3), dtype=np.uint8)
    
    for i in range(altura):
        degradado[i, :] = [int(255*i/altura), int(255*(altura-i)/altura), 255]

   
    plt.imshow(degradado)
    plt.axis('off')
    plt.show()

altura = 200
ancho = 800
generar_degradado_lineal(altura, ancho)

```

Ejercicio 3 

```python
import numpy as np
import matplotlib.pyplot as plt

def generar_tablero_ajedrez(filas, columnas, tamao_casilla):
    tablero = np.zeros((filas*tamao_casilla, columnas*tamao_casilla, 3), dtype=np.uint8)

    for fila in range(filas):
        for columna in range(columnas):
            if (fila + columna) % 2 == 0:
                tablero[fila*tamao_casilla:(fila+1)*tamao_casilla, columna*tamao_casilla:(columna+1)*tamao_casilla] = [255, 255, 255]

    plt.imshow(tablero)
    plt.axis('off')
    plt.show()

filas = 8
columnas = 8
tamao_casilla = 100
generar_tablero_ajedrez(filas, columnas, tamao_casilla)
```

Ejercicio 4

```python
import cv2
import matplotlib.pyplot as plt

img_original = cv2.imread('imagen1.jpg')
img_recortada = cv2.imread('Imagenrecortada.jpg')

img_recortada_gris = cv2.cvtColor(img_recortada, cv2.COLOR_BGR2GRAY)

img_ecualizada = cv2.equalizeHist(img_recortada_gris)

hist_original = cv2.calcHist([img_recortada_gris], [0], None, [256], [0, 256])
hist_ecualizada = cv2.calcHist([img_ecualizada], [0], None, [256], [0, 256])

plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(img_recortada_gris, cmap='gray')
plt.title('Imagen Recortada (Escala de Grises)')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(img_ecualizada, cmap='gray')
plt.title('Imagen Ecualizada')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.plot(hist_original, color='black')
plt.plot(hist_ecualizada, color='red')
plt.title('Histogramas')
plt.xlabel('Intensidad de pxel')
plt.ylabel('Frecuencia')
plt.legend(['Original', 'Ecualizada'])

plt.show()
```

Ejercicio 5

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

imagen_original = cv2.imread('imagen1.jpg')

altura, ancho, _ = imagen_original.shape

imagen_degradada = np.zeros_like(imagen_original)
for i in range(altura):
    for j in range(ancho):
        valor = int(255 * j / ancho)
        imagen_degradada[i, j] = [valor, valor, valor]

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imagen_original, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(imagen_degradada, cv2.COLOR_BGR2RGB))
plt.title('Degradado Lineal')
plt.axis('off')

plt.show()
```
